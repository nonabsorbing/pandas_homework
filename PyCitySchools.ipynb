```python
# Dependencies and Setup
import pandas as pd
import numpy as np


# File to Load (Remember to Change These)*******!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"


# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_oldheads = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

# Check it. 
school_data_oldheads.head()
```

```python
#Rename headers - create default dataframe 

schooldata = school_data_oldheads.rename(columns={"student_name":"Student Name", "gender" : "Gender", "grade":"Grade", "school_name":"School Name", "reading_score": "Reading Score", "math_score": "Math Score", "type":"School Type", "size" : "Number of Students", "budget" :"School Budget"})

schooldata.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Keep one row for each school name (since a lot of our data is already in there)
district_summ = schooldata.drop_duplicates("School Name")

#Number of schools = length of unique rows after reducing duplicats
schools_number = len(district_summ)

#pull budget number for output
district_budget = district_summ["School Budget"].sum()
    # district_budget = (district_budget(${:,.2f}'.)format

#pull students# for output
district_students = district_summ["Number of Students"].sum()

#math average = sum of math scores divided by # of students (length of dataset)
district_mathavg = (schooldata["Math Score"].sum())/(len(schooldata))

#reading average = sum of reading scores divided by # of students (length of dataset)
district_readavg = (schooldata["Reading Score"].sum())/(district_students)


#overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
district_passrate = (district_readavg + district_mathavg)/2
    
#Calculate the percentage of students with a passing math score (70 or greater)

#Get number of math scores above 70
district_mathpass = schooldata.loc[schooldata['Math Score'] >= 70]

#get number of reading scores above 70
district_readpass = schooldata.loc[schooldata['Reading Score'] >= 70]

#Calculate the percentage of students with a passing math score (70 or greater)
passing_mathcount = district_mathpass["Math Score"].count
passing_mathpercent = district_mathpass["Math Score"].sum()/passing_mathcount()

#Calculate the percentage of students with a passing reading score (70 or greater)
passing_readcount = district_readpass["Reading Score"].count
passing_readpercent = district_readpass["Reading Score"].sum()/passing_readcount()
```

```python
#make dataframe for district data 

district_summary = pd.DataFrame({"Number of Schools" : schools_number, 
                                 "District Budget": district_budget, 
                                 "Number of Students": district_students, 
                                 "Average Math Score" : district_mathavg, 
                                 "Average Reading Score" : district_readavg, 
                                 "District Passing Rage" : district_passrate,
                                 "Percent of Students Passing Math" : passing_mathpercent,  
                                 "Percent of Students Passing Reading" : passing_readpercent                                  
                                }, index=["District Summary"])


district_summary.head()

#*****************GO BACK AND FORMAT! change index to # of schools 
#Take out zero?? 
```
